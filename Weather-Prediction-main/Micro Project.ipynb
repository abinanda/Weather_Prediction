#### Categorical values to numerical values

```python
def LABEL_ENCODING(c1):
    from sklearn import preprocessing
    label_encoder = preprocessing.LabelEncoder()
    df[c1]= label_encoder.fit_transform(df[c1])
    df[c1].unique()
    print(df[c1].unique)
LABEL_ENCODING("weather")
```

```python
df
```

```python
df.weather.value_counts()
```

```python
df = df.drop('date',axis=1)
```

```python
X = df.drop('weather',axis=1)
y = df['weather']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train_ss = ss.fit_transform(X_train)
X_test_ss = ss.transform(X_test)
```

#### LOR

```python
from sklearn.linear_model import LogisticRegression
LOR = LogisticRegression(random_state=0)
LOR.fit(X_train_ss, y_train)
```

```python
y_pred_LOR = LOR.predict(X_test)
y_pred_LOR
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred_LOR)
print(cm)
```

```python
sns.heatmap(cm,annot=True)
plt.show()
```

```python
acc1 = accuracy_score(y_test, y_pred_LOR)
print(f"Accuracy score: {acc1}")
```

```python
y_pred = LinSVC.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
acc2 = accuracy_score(y_test, y_pred)
```
